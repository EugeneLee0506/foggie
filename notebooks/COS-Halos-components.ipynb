```python
from astropy.io import fits

import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'stixgeneral'
mpl.rcParams['font.size'] = 16.
%matplotlib notebook
```

```python
data = fits.open('/Users/molly/Dropbox/foggie-collab/cos-halos-ioncomponentstats.fits')
data[1].data
```

```python
si2color = "#984ea3"
si3color = "#4575b4"
c3color = 'darkorange'
o6color = "#4daf4a"

fig = plt.figure(figsize=(6,6))

plt.scatter(data[1].data['rho'],data[1].data['siiiincomp'],marker='D',lw=2,facecolors='none',edgecolors=si3color,label='Si III')
plt.scatter(data[1].data['rho'],data[1].data['ciiincomp'],marker='*',lw=2,facecolors='none',edgecolors=c3color,label='C III')
plt.scatter(data[1].data['rho'],data[1].data['ovincomp'],marker='o',lw=2,facecolors='none',edgecolors=o6color,label='O VI')
plt.scatter(data[1].data['rho'],data[1].data['siiincomp'],marker='s',lw=2,facecolors='none',edgecolors=si2color,label='Si II')

plt.ylim(0.1,5.2)
plt.xlabel('impact parameter [kpc]',fontsize=18)
plt.ylabel('number of components',fontsize=18)

plt.legend()
```

```python
fig = plt.figure(figsize=(6,6))

plt.scatter(data[1].data['rho']/data[1].data['rvir'],data[1].data['siiiincomp'],marker='D',lw=2,facecolors='none',edgecolors=si3color,label='Si III')
plt.scatter(data[1].data['rho']/data[1].data['rvir'],data[1].data['ciiincomp'],marker='*',lw=2,facecolors='none',edgecolors=c3color,label='C III')
plt.scatter(data[1].data['rho']/data[1].data['rvir'],data[1].data['ovincomp'],marker='o',lw=2,facecolors='none',edgecolors=o6color,label='O VI')
plt.scatter(data[1].data['rho']/data[1].data['rvir'],data[1].data['siiincomp'],marker='s',lw=2,facecolors='none',edgecolors=si2color,label='Si II')

plt.xlabel(r'impact parameter / R$_{vir}$', fontsize=18)
plt.ylabel('number of components', fontsize=18)
plt.ylim(0.1,5.2)

plt.legend()
```
