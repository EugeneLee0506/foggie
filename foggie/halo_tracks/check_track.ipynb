```python
import numpy as np

from astropy.table import Table
from astropy.io import ascii

import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.size'] = 16.

%matplotlib notebook
```

```python
track = Table.read('nref11n_selfshield_15/track.dat', format='ascii.fixed_width_two_line')
```

```python
fig = plt.figure(figsize=(9,7))
ax = fig.add_subplot(111)
ax.step(track['redshift'], track['x0'])
plt.xlim((1.95, 6.2))
```

```python
fig = plt.figure(figsize=(9,7))
ax = fig.add_subplot(111)
ax.step(track['redshift'], track['x0'], label="x")
ax.step(track['redshift'], track['y0'], label="y")
ax.step(track['redshift'], track['z0'], label="z")
plt.legend()
plt.xlim((1.95, 6.2))
```

```python
plt.step(track['redshift'], track['z0'])
plt.xlim((1.95,6.2))
```

```python
## try to calculate the difference
dr = np.zeros(len(track))
for i in range(len(track)-1):
    dr[i] = np.sqrt((track['x0'][i+1]-track['x0'][i])**2 
                 + (track['y0'][i+1]-track['y0'][i])**2 
                 + (track['z0'][i+1]-track['z0'][i])**2)
```

```python
fig = plt.figure(figsize=(9,7))
ax = fig.add_subplot(111)
ax.step(track['redshift'], dr*100*1000)
plt.ylim(0,5)
plt.xlim((1.9, 6))
```

```python
new_track = Table.read('nref11n_selfshield_15/track_temp.dat', format='ascii.fixed_width_two_line')
new_dr = np.zeros(len(new_track))
for i in range(len(new_track)-1):
    new_dr[i] = np.sqrt((new_track['x0'][i+1]-new_track['x0'][i])**2 
                 + (new_track['y0'][i+1]-new_track['y0'][i])**2 
                 + (new_track['z0'][i+1]-new_track['z0'][i])**2)
```

```python
fig = plt.figure(figsize=(9,7))
ax = fig.add_subplot(111)
ax.step(new_track['redshift'], new_dr*100*1000)
plt.ylim(0,5)
plt.xlim((3, 6.2))
```
