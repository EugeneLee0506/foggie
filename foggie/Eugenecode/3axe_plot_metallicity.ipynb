```python
import numpy as np
import matplotlib.pyplot as plt
home_dir = '/Users/hylee/foggie/'
output_dir = home_dir + 'foggie_output'
```

```python
y_red = np.load('met_y_redshift.npy')
y_time = np.load('met_y_time.npy')
y_cutoff = np.load('met_y_cutoff.npy')
```

```python
x_red = np.load('met_x_redshift.npy')
x_time = np.load('met_x_time.npy')
x_cutoff = np.load('met_x_cutoff.npy')
z_red = np.load('met_z_redshift.npy')
z_time = np.load('met_z_time.npy')
z_cutoff = np.load('met_z_cutoff.npy')
```

```python
#plt.subplot(1, 2, 1)
fig=plt.figure()
plt.plot(x_red, x_cutoff, 'ro-', label = "x axis projection")
plt.plot(y_red, y_cutoff, 'go-', label = "y axis projection")
plt.plot(z_red, z_cutoff, 'bo-', label = "z axis projection")
plt.title('Metallicity', fontsize = 20)
plt.xlabel('Redshift [z]', fontsize = 15)
plt.ylabel('Metallicity turnoff radius [kpc]', fontsize = 15)
plt.legend(loc="upper left", fontsize = 17.5)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.tight_layout()
plt.ylim(0, 20)
plt.gca().invert_xaxis()

plt.show()
fig.savefig(output_dir + "pptt.jpg", bbox_inches = 'tight')
```
