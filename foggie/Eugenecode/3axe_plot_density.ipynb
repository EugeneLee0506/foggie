```python
import numpy as np
import matplotlib.pyplot as plt
home_dir = '/Users/hylee/foggie/'
output_dir = home_dir + 'foggie_output'
```

```python
y_red = np.load('y_axis_redshift.npy')
y_time = np.load('y_axis_timelist.npy')
y_cutoff = np.load('y_axis_cutofflist.npy')
```

```python
x_red = np.load('x_axis_redshiftlist.npy')
x_time = np.load('x_axis_timelist.npy')
x_cutoff = np.load('x_axis_cutofflist.npy')
z_red = np.load('z_axis_redshift.npy')
z_time = np.load('z_axis_timelist.npy')
z_cutoff = np.load('z_axis_cutofflist.npy')
```

```python
#plt.subplot(1, 2, 1)
fig=plt.figure()
plt.plot(x_time, x_cutoff, 'ro-', label = "x axis projection")
plt.plot(y_time, y_cutoff, 'go-', label = "y axis projection")
plt.plot(z_time, z_cutoff, 'bo-', label = "z axis projection")
plt.title('Density', fontsize = 15)
plt.xlabel('Time [Gyr]', fontsize = 15)
plt.ylabel('Disk size [kpc]', fontsize = 15)
plt.legend(loc="upper left", fontsize = 17.5)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.tight_layout()
plt.show()
fig.savefig(output_dir + "density.jpg", bbox_inches = 'tight')
```

```python
#plt.subplot(1, 2, 2)
fig2=plt.figure()
plt.plot(x_red, x_cutoff, 'ro-', label = "x axis projection")
plt.plot(y_red, y_cutoff, 'go-', label = "y axis projection")
plt.plot(z_red, z_cutoff, 'bo-', label = "z axis projection")
plt.title('Density', fontsize = 20)
plt.xlabel('Redshift [z]', fontsize = 15)
plt.ylabel('Disk size [kpc]', fontsize = 15)
plt.legend(loc="upper left", fontsize = 17.5)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.gca().invert_xaxis()
plt.tight_layout()
plt.show()

fig2.savefig(output_dir + "3axes_cutoff_redshift.jpg", bbox_inches = 'tight')
```
