```python
x = np.random.normal(4, 3, (3, 100000))

theta = np.arccos(x[2]/np.sqrt(sum(x**2., axis = 0)))*180./pi
phi = np.arctan(x[1]/x[0])*180./pi

fig, ax = plt.subplots(1,1, figsize = (5,5))
ax.hist2d(phi, theta, bins = 100)
ax.set_xlim(-90, 90)
ax.set_ylim(0, 180)

fig.set_dpi(400)
```
